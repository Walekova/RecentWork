```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
import ibm_boto3
from ibm_botocore.client import Config, ClientError
import time
import numpy as np
from io import BytesIO
import urllib.request as req
import wget
from PIL import Image
from datetime import datetime
import io as libio
import sched, time
from datetime import datetime, timedelta
from threading import Timer
import xml.dom.minidom
from xml.etree import ElementTree
import os
file_path = os.getcwd()

s = sched.scheduler(time.time, time.sleep)
       
def writeImage(tod):
    
    cities_df = pd.read_csv('cities_allinfo.csv')
    
    data_col = ['key','city', 'state', 'lat', 'lon', 'time', 'timeZone', 'obsDaylight', 'currentGmtOffset', 'timeZoneAbbreviation', \
       'observationtime', 'pressurestate', 'temperature', 'realfeel', 'humidity', 'weathertext', 'weathericon', 'windgusts', \
       'windspeed', 'winddirection', 'visibility', 'precip', 'uvindex', 'dewpoint', 'cloudcover']

    weather_24 = pd.DataFrame(columns=data_col)
    
    for index, row in cities_df.iterrows():
        get_locationKey = row['Key']
        print(get_locationKey)
        current = requests.get("http://samsungmobile.accu-weather.com/widget/samsungmobile/weather-data.asp?metric=1&location=cityId%3A"+ str(get_locationKey))
        
        root = ElementTree.fromstring(current.content)
    
        conditions = [[get_locationKey ,root[1][0].text, root[1][1].text, root[1][2].text, root[1][3].text, root[1][4].text, root[1][5].text, \
                  root[1][6].text, root[1][7].text, root[1][8].text, root[2][1].text, root[2][2].text, root[2][3].text, \
                  root[2][4].text, root[2][5].text, root[2][6].text, root[2][7].text, root[2][8].text, root[2][9].text, \
                  root[2][10].text, root[2][11].text, root[2][12].text, root[2][13].text, root[2][14].text, root[2][15].text]]
    
        current_df = pd.DataFrame(conditions, columns=data_col)
        weather_24 = pd.concat([weather_24, current_df])
    
    file = "EU_" + str(tod) + ".csv"
    weather_24.to_csv(file)
    weather_24.drop(weather_24.index, inplace=True)
    
def do_something(sc): 
    # do your stuff
    dt_object = datetime.now()
    check = dt_object.strftime('%M')
    if check == '45':
        tod = dt_object.strftime('%y%m%d%H%M')
        writeImage(tod)
        
    s.enter(60, 1, do_something, (sc,))

s.enter(60, 1, do_something, (s,))
s.run()
```
